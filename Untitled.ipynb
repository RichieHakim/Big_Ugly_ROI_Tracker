```python
## Initialize the similarity matrix
def make_similarity_matrix(data):
    d = torch.cdist(data, data).type(torch.float32)  ## Get distance matrix
    s = torch.maximum(torch.log(1/d), torch.zeros_like(d))  ## Convert to similarity matrix and relu. Distances > 1 end up as zero
    s[torch.eye(s.shape[0]).type(torch.bool)] = 0  ## Set identity similarities to zeros (this helps later when taking sums)
    s = s.to_sparse_coo()  ## Make similarity matrix a sparse tensor
    return s
s = make_similarity_matrix(data)
```

```python
## Initialize the multihot matrix
def initialize_multihot_matrix(n_samples, n_clusters):
    h = torch.rand(size=(n_samples, n_clusters))  ## Random initialization
    h[:,0] = torch.cat((torch.ones(100), torch.zeros(100), torch.ones(300), torch.zeros(500)))  ## funny business
    h = torch.nn.functional.softmax(h, dim=1)  ## Softmax. not sure why yet
    return h
h = initialize_multihot_matrix(n_samples=n_samples, n_clusters=200)
```
